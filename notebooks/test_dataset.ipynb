## Implement helper for dataset visualization

```python
def vizualize_dataset(dataset):
    fig, axs = plt.subplots(2, 2, figsize=(8, 8))
    fig.tight_layout()
    
    angles, images = dataset['angles'], dataset['images']
    axs[0][0].imshow(images[0], cmap="Greys")
    axs[0][0].title.set_text("First image from the dataset")

    axs[0][1].imshow(images[-1], cmap="Greys")
    axs[0][1].title.set_text("Last image from the dataset")

    axs[1][0].scatter(angles[:,0], angles[:,1], 1)
    axs[1][0].title.set_text("Distribution of angles")
    
    axs[1][1].hist(x_train.flatten())
    axs[1][1].title.set_text("Distribution of pixel values")
```
